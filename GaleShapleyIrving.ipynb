# 1. Find the best and worst cases for the Gale-Shapley and Irving ( An Efficient Algorithm for the “Stable Roommates” Problem) Algorithms and run in Colab these algorithms for those cases for n=4 and 40.

```
def stable_marriage(n, m_prefer, w_prefer):

  bachelors = set([i for i in range(n)])
  couples = [-1 for i in range(n)]
  count = 0

  while (len(bachelors) > 0):

    b = bachelors.pop()

    for w in m_prefer[b]:
      count += 1

      if couples[w] == -1 :
        couples[w] = b
        break
      else :
        if w_prefer[w].index(b) < w_prefer[w].index(couples[w]):
          bachelors.add(couples[w])
          couples[w] = b
          break
  return count, couples
```

```
def get_best_case(n):
  m = []
  w = []
  for i in range(n):
    m.append([])
    w.append([])
    m[i].append(i)
    w[i].append(i)
    for j in range(n):
      if i != j:
        m[i].append(j)
        w[i].append(j)
  return n, m, w
```

```
def get_worst_case(n):
  m = []
  w = []
  for i in range(n):
    m.append([])
    w.append([])
    for j in range(n):
      m[i].append(j)
      w[i].append(n - j - 1)
  return n, m, w
```

## n = 4

```
n = 4
case = get_best_case(n)
count, couples = stable_marriage(case[0], case[1], case[2])
print("Best case for n: ", n)
print("Man prefer")
for i in range(n):
  print(i, "\t", case[1][i])
print("Woman prefer")
for i in range(n):
  print(i, "\t", case[2][i])
print("Number of iterations: ", count)
print("Woman ", " Man") 
for i in range(n): 
    print(i, "\t", couples[i])
case = get_worst_case(n)
count, couples = stable_marriage(case[0], case[1], case[2])
print("Worst case for n: ", n)
print("Man prefer")
for i in range(n):
  print(i, "\t", case[1][i])
print("Woman prefer")
for i in range(n):
  print(i, "\t", case[2][i])
print("Number of iterations: ", count)
print("Woman ", " Man") 
for i in range(n): 
    print(i, "\t", couples[i])
```

# 2. Run the Gale-Shapley Algorithm 100 times using random permutations as preference lists for n=4 and 40. Find the minimum, maximum and average number of iterations and plot the distribution of the number of iterations.

```
def get_random_case(n):
  m = []
  w = []
  for i in range(n):
    m.append([])
    w.append([])
    for j in random.permutation(np.array([i for i in range(n)])):
      m[i].append(j)
    for j in random.permutation(np.array([i for i in range(n)])):
      w[i].append(j)
  return n, m, w
```

## n = 4

```
it_num = []
for i in range(100):
  case = get_random_case(4)
  it_num.append(stable_marriage(case[0], case[1], case[2])[0])
min = np.amin(it_num)
print("Minimum", min)
max = np.amax(it_num)
print("Maximum", max)
print("Average", np.mean(it_num))
plt.hist(it_num, bins = [i for i in range(min, max + 1)])
plt.show()
```

## n = 40

```
it_num = []
for i in range(100):
  case = get_random_case(40)
  it_num.append(stable_marriage(case[0], case[1], case[2])[0])
min = np.amin(it_num)
print("Minimum", min)
max = np.amax(it_num)
print("Maximum", max)
print("Average", np.mean(it_num))
plt.hist(it_num, bins = [i for i in range(min, max + 1)])
plt.show()
```

# 3. (Extra credit) Simulate the the Gale-Shapley and Irving algorithms using random preference lists , for n=4 and 40 and generate a gift animation fo the type show in “Illustrating the Deferred Acceptance Algorithm with R – Code https://www.r-bloggers.com/illustrating-the-deferred-acceptance-algorithm-with-r/

```
def stable_marriage_images_for_animation(n, m_prefer, w_prefer):

  bachelors = set([i for i in range(n)])
  couples = [-1 for i in range(n)]
  count = 0
  images = []

  while (len(bachelors) > 0):

    b = bachelors.pop()

    for w in m_prefer[b]:
      images.append([])
      for i in range(n):
        images[count].append([])
        for j in range(n):
          if b == i and w == j and couples[j] == i:
            images[count][i].append(64)
          elif couples[j] == i:
            images[count][i].append(0)
          elif b == i and w == j :
            images[count][i].append(128)
          else:
            images[count][i].append(255)
      count += 1
      if couples[w] == -1 :
        couples[w] = b
        break
      else :
        if w_prefer[w].index(b) < w_prefer[w].index(couples[w]):
          bachelors.add(couples[w])
          couples[w] = b
          break
  return count, images
```

```
iter=0
def generate_animation(count, images):
  def updatefig(*args):
    global iter
    iter += 1
    iter %= count
    im.set_array(images[iter])
    return im,

  fig = plt.figure()
  plt.xlabel('women')
  plt.ylabel('men')
  ax = plt.gca()
  ax.invert_yaxis()
  ax.yaxis.set_major_locator(MaxNLocator(integer=True))
  im = plt.imshow(images[0], cmap='gray', vmin=0, vmax=255, animated=True)
  return animation.FuncAnimation(fig, updatefig, save_count = count, interval = 500)
```

```
case = get_random_case(4)
count, images = stable_marriage_images_for_animation(case[0], case[1], case[2])
anim = generate_animation(count, images)
anim
```
